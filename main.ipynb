```python
import pandas as pd
import numpy as np

credit = pd.read_csv('movies_db/tmdb_5000_credits.csv')
movies = pd.read_csv('movies_db/tmdb_5000_movies.csv')
credit
```

```python
movies
```

```python
movies = movies.merge(credit, on='title')
```

```python
movies.shape
```

```python
movies['original_language'].value_counts()
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.reset_index(drop= True, inplace= True)
```

```python
movies = movies.drop_duplicates(subset='movie_id')
movies.reset_index(drop=True, inplace=True)

print("movies:", len(movies))
```

```python
movies.iloc[0]['genres']
```

```python
import ast

def convert(obj):
    List = []
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert_name(obj):
    List = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3:
            List.append(i['name'])
            count += 1
        else:
            break
    return List
```

```python
movies['cast'] = movies['cast'].apply(convert_name)
```

```python
def fetch(obj):
    List = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            List.append(i['name'])
            break
    return List
```

```python
movies['crew'] = movies['crew'].apply(fetch)
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags', 'overview']]
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : " ".join(x).lower())
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
ps.stem
```

```python
def steming(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
new_df['tags'] = new_df['tags'].apply(steming)
new_df['overview'] = new_df['overview'].apply(lambda x: ' '.join(x))
new_df.reset_index(drop= True, inplace= True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features= 5000, stop_words= 'english')
vectors = cv.fit_transform(new_df['tags']).toarray()

```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors).astype(np.float32)
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])), reverse= True, key= lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse= True, key= lambda x: x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
import pickle
with open('movies.pkl', 'wb')as f:
    pickle.dump(new_df, f)

with open('similarity.pkl', 'wb') as f:
    pickle.dump(similarity, f)
```
